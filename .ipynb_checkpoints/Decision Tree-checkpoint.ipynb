```python
train = pd.read_csv('train.csv')
df =  pd.read_csv('converted1.csv')
df = df.append(pd.read_csv('converted2.csv'),ignore_index=True)
df = df.append(pd.read_csv('converted3.csv'),ignore_index=True)
df = df.append(pd.read_csv('converted4.csv'),ignore_index=True)
df = df.append(pd.read_csv('converted5.csv'),ignore_index=True)
df = df.append(pd.read_csv('converted6.csv'),ignore_index=True)
len(df)-len(train)
```

```python
#train.reset_index(inplace=True)
#train.drop(columns='index',inplace=True)
final = train.merge(df,left_index=True,right_index=True)
final.head()
```

```python
len(final)
```

```python
final['n_sex']=0
final['n_sex'][final['sex']=='male']=1
final.head()
```

```python
final['anatom_site_general_challenge'].value_counts()
```

```python
final.rename(columns={'anatom_site_general_challenge':'site'},inplace=True)
final['torso']=0
final['torso'][final['site']=='torso']=1
final['l_extremity']=0
final['l_extremity'][final['site']=='lower extremity']=1
final['u_extremity']=0
final['u_extremity'][final['site']=='upper extremity']=1
final['head/neck']=0
final['head/neck'][final['site']=='head/neck']=1
final['palms/soles']=0
final['palms/soles'][final['site']=='palms/soles']=1
final.head()
```

```python
final2 = final.drop(columns=['image_name','patient_id','sex','site','diagnosis','benign_malignant'])
final2.dropna(inplace=True)
final2.head()
```

```python
final2.isna().sum()
```

```python
X_train, X_test, y_train, y_test = train_test_split(final2.drop(columns='target'), final2['target'], test_size=0.2) 
D_tree.fit(X_train,y_train)
```

```python
# Code here
train_acc = D_tree.score(X_train,y_train)
test_acc = D_tree.score(X_test,y_test)

print(f'The accuracty of the model on the training set was {train_acc}, \
while the accuracy of the model on the testing was only {test_acc}')
```

```python
features = final2.drop(columns='target').columns
#D_tree.feature_impontances_
pair_df = pd.DataFrame(data=[features,D_tree.feature_importances_])
pair_df = pair_df.transpose()
pair_df.sort_values(1,inplace=True)
plt.bar(pair_df[0],pair_df[1])
plt.xticks(rotation=35)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline


# Code here
min_samples_split_values = [3,4,8,16,32,64,128,256,512,1024]
min_samples_leaf_values = [2,3,4,5,10,20,25,50,100,250]

# Code here
scores=[]
for x in min_samples_split_values:
    for y in min_samples_leaf_values:
        score =[]
        D_tree = DecisionTreeClassifier(criterion='entropy', min_samples_split=x, min_samples_leaf=y)
        D_tree.fit(X_train, y_train)
        test_acc = D_tree.score(X_test,y_test)
        score = [x,y,test_acc]
        scores.append(score)
df2 = pd.DataFrame(scores, columns = ['min_samples_split', 'min_samples_leaf','accuracy'])

fig = plt.figure(figsize = (12, 6))
ax = plt.subplot(111)
for y in min_samples_leaf_values:
    c = (np.random.rand(), np.random.rand(), np.random.rand())
    data = df2[df2['min_samples_leaf']==y]
    data.sort_values(by='min_samples_split',inplace=True)
    plt.plot(data['min_samples_split'],data['accuracy'],color=c,label=f'min_sample_leaf={y}')

       
# Put a legend below current axis
box = ax.get_position()
ax.set_position([box.x0, box.y0 + box.height * 0.0 , box.width, box.height * 1])
ax.legend(loc = 'upper center', bbox_to_anchor = (0.5, -0.15), fancybox = True, 
              shadow = True, ncol = 4, prop = {'size':10})
ax.set_xlabel('min_sample_split')
```
